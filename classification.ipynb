```python
def sigmoid(z):
    if z : 
        try:
            z = np.clip(z,-300,300)
        
        except Exception as e: 
            print(str(e))

    return 1 /(1 + np.exp(-z))
```

```python
def normalizedata(Train,Test):
    train_mean = np.mean(Train, axis=0)
    train_std = np.std(Train,axis=0)

    Train = (Train - (train_mean/train_std))
    Test = (Test - (train_mean/train_std))

    return Train,Test

```
